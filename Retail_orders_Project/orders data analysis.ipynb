```python
# import libraries for zipfile and unzip the downloaded file
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() #extract the file to dir
zip_ref.close() #close file. 
```

```python
# read data from file and handle null values
import pandas as pd
df= pd.read_csv('orders.csv')
df.head()
```

```python
# to find the unique/distinct values from the column - ship mode
df['Ship Mode'].unique()
```

```python
# make the not available/unknown values to null 
df= pd.read_csv('orders.csv',na_values =['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename column names - make them lower case
#df.rename(columns={'Order Id':'order_id','City':'city'}) -- this is one way of doing 
df.columns = df.columns.str.lower()
df
```

```python
#rename column names - remove space and place underscore 
df.columns = df.columns.str.replace(' ','_')
df
```

```python
#derive new columns discount, sale price and profit 
#discount
df['discount'] = df['list_price']*df['discount_percent']/100
#sale price 
df['sale_price'] = df['list_price']-df['discount']
#total profit 
df['profit'] = df['sale_price'] -df['cost_price']
df.head(5)
```

```python
# convert order date from object to datetime
# df.dtypes to check the datatypes of each column 
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
#drop the columns that are not needed for analysis 
df.drop(columns=['list_price','cost_price','discount_percent'], inplace = True)
```
